Lets chcek the major reason of kidnapping in each and every state

```python
#importing the dataset
import pandas as pd
df_kidnap=pd.read_csv("39_Specific_purpose_of_kidnapping_and_abduction.csv")
df_kidnap
```

```python
#lets group by area and reason and count the total number of kidnaps
df_group=df_kidnap.groupby(["Area_Name","Group_Name"]).agg({"K_A_Grand_Total":"sum"})
```

```python
#lets display the dataset
df_group.loc[("Andaman & Nicobar Islands")]
```

```python
#lets do the find the major reson of kidnapping in each state
state_list=df_kidnap["Area_Name"].drop_duplicates().to_list()
#Lets make a table which consist of AREA major reason of kidnapping and count of kidnaps_major reason,total kidnapping and percentage of kidnnaps due to major reason
df_major_reason_kidnap=pd.DataFrame(columns=["AREA","Major_reason_of_kidnap","Count_of_kidnaps_due_to_major_reason","Total_kidnaps","Proportion_of_kidnaps_major_reason"])
area_list=list()
major_reason_list=list()
count_kidnaps_major=list()
total_kidnaps=list()
proportion_list=list()
for state in state_list:
    max_kid=0
    for index in df_group.loc[state].index:
        #print(df_group.loc[state].loc[index][0])
        if (max_kid<df_group.loc[state].loc[index][0]) and (index!="Kidnap - Total"):
            max_kid=df_group.loc[state].loc[index][0]
            index_final=index
            #print(max_kid)
    print("AREA: {} \nMajority cases due to: {} and count is: {}".format(state,index_final,max_kid))
    print("Ratio of cases due to {} is :{}".format(index_final,max_kid/df_group.loc[state].loc["Kidnap - Total"][0]))
    print("\n")
    area_list.append(state)
    major_reason_list.append(index_final)
    count_kidnaps_major.append(max_kid)
    total_kidnaps.append(df_group.loc[state].loc["Kidnap - Total"][0])
    proportion_list.append(max_kid/df_group.loc[state].loc["Kidnap - Total"][0])
```

```python
# Displaying table for major reason people being kidnapped in each and every state
df_major_reason_kidnap["AREA"]=area_list
df_major_reason_kidnap["Major_reason_of_kidnap"]=major_reason_list
df_major_reason_kidnap["Count_of_kidnaps_due_to_major_reason"]=count_kidnaps_major
df_major_reason_kidnap["Total_kidnaps"]=total_kidnaps
df_major_reason_kidnap["Proportion_of_kidnaps_major_reason"]=proportion_list
df_major_reason_kidnap
```

OFFENDER's relation to rape victim

```python
#importig the dataset
df_rape=pd.read_csv("20_Victims_of_rape.csv")
df_rape
```

```python
#lets chcek the total of rape victims grouped by AREA and Subgroup
df_rape_group=df_rape.groupby(["Area_Name","Subgroup"]).agg({"Victims_of_Rape_Total":"sum"})
```

```python
#displaying the sum as per groups
df_rape_group
```

```python
#Lets make a table showing total sum of incest rape cases in each state
df_rape_incest=pd.DataFrame(columns=["Area","incest_rape","total_rape"])
area_list=df_rape["Area_Name"].drop_duplicates().to_list()
incest_rape=list()
total_rape=list()
for area in area_list:
    incest_rape.append(df_rape_group.loc[area].loc["Victims of Incest Rape"][0])
    total_rape.append(df_rape_group.loc[area].loc["Total Rape Victims"][0])
df_rape_incest["Area"]=area_list
df_rape_incest["incest_rape"]=incest_rape
df_rape_incest["total_rape"]=total_rape
```
