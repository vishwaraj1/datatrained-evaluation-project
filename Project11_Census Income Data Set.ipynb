```python
# Replace ? with Unknown
for col in df.columns:
    df.loc[df[col] == '?', col] = 'unknown'
```

```python
# Check if ? is present
for col in df.columns:
    print(f"? in {col}: {df[(df[col] == '?')].any().sum()}")
```

```python
#lets convert catorigical data in numerical
from sklearn.preprocessing import LabelEncoder
lbe=LabelEncoder()
for col in df.columns:
    if df[col].dtypes=="object":
        df[col]=lbe.fit_transform(df[col])
```

```python
df.head()
```

```python
#lets divide dataset into input and output 
x=df.drop(['Target'],axis=1)
x.head(3)
```

```python
y=df[['Target']]
y.head(4)
```

```python
#lets bring down data in scale
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import f1_score,roc_auc_score
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x, y,random_state = 42,test_size=0.20,stratify=y)
model=[LogisticRegression(),SVC(),KNeighborsClassifier(),GaussianNB(),XGBClassifier(),
       DecisionTreeClassifier(),RandomForestClassifier()]
```

```python
#lets check which model perform best
import warnings
warnings.filterwarnings("ignore")
for m in model:
    m.fit(x_train,y_train)
    pred=m.predict(x_test)
    print("model",m, "aucroc_score",roc_auc_score(y_test,pred))
    print()
```

```python
from imblearn.over_sampling import SMOTE
df_x,y = SMOTE().fit_sample(x, y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = 76,test_size=0.20,stratify=y)
x_train, y_train = SMOTE().fit_sample(x_train, y_train)
rf_clf=RandomForestClassifier()
rf_clf.fit(x_train,y_train)
y_pred=rf_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
print("Confusion matrix \n",confusion_matrix(y_test,y_pred))
print("f1 score is : ",f1_score(y_test,y_pred))
print("classification report \n",classification_report(y_test,y_pred))
print("AUC ROC Score: ",roc_auc_score(y_test,y_pred))
```
