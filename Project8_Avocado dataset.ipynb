```python
# Date is Object type not the date type. We have to change its type to date type.
df['Date']=pd.to_datetime(df['Date'])
df['Month']=df['Date'].apply(lambda x:x.month)
df['Day']=df['Date'].apply(lambda x:x.day)
```

```python
df.drop(columns=['Date'],axis=1,inplace=True)
```

```python
# As we can see we have 54 regions and 2 unique types, so it's going to be easy to to transform the typefeature to dummies, but for the region its going to be a bit complex, so I decided to drop the entire column.
#df1=pd.get_dummies(df.drop(['region'],axis=1),drop_first=True)
```

```python
df = pd.get_dummies(df, columns=['type'],drop_first=True)
```

```python
df.head()
```

```python
#label encode region
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['region'] = le.fit_transform(df['region'])
df.head()
```

```python
x=df.drop(columns='AveragePrice')
y=df[['AveragePrice']]
```

```python
x.skew()
```

```python
#lets treat the skewness in the input dataset
import numpy as np
for index in x.skew().index:
    if x.skew().loc[index]>0.5:
        x[index]=np.log1p(x[index])
   # if x.skew().loc[index]<-0.5:
        #x[index]=np.square(x[index])
```

```python
x.skew()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = 4,test_size=0.30)
```

```python
model=[LinearRegression(),AdaBoostRegressor(),GradientBoostingRegressor(),SVR(),KNeighborsRegressor(),XGBRegressor(),RandomForestRegressor()]
for m in model:
    m.fit(x_train,y_train)
    pred=m.predict(x_test)
    print('model',m,'r2 score is:',r2_score(y_test,pred))
    print()
    print()
```

```python
from sklearn.ensemble import RandomForestRegressor
x_train,x_test,y_train,y_test=train_test_split(x, y,random_state = 81,test_size=0.20)
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
```

```python
#Lets find the rmse and r2_score using sklearn.metrics for random forest model
from sklearn.metrics import r2_score , mean_squared_error
print("MSE is: ",mean_squared_error(y_test,y_pred))
print("r2_score is: ",r2_score(y_test,y_pred))
```
